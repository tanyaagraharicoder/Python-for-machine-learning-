Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



```

```python
var= sns.load_dataset("penguins")
var
```

```python
sns.set(style="dark")
order_1=["Dream", "Torgersen","Biscoe"]
sns.barplot(x=var.island, y=var.bill_length_mm, data=var  , hue="sex",
           order=order_1, hue_order=["Female", "Male"], orient="v" , palette="icefire",
            saturation=0.9 ,capsize=1 , dodge=True, alpha=0.5 ,  )
plt.show()
```

```python

sns.barplot(x="bill_length_mm", y="bill_depth_mm", data=var, orient="v")
plt.show()
```
